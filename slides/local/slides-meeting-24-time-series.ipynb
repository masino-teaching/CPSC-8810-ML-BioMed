```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima_process import ArmaProcess

# Simulate a time series using an ARMA model with 1 AR and 1 MA component
np.random.seed(42) # For reproducibility
ar = np.array([1, -0.5]) # AR parameter
ma = np.array([1, 0.5])  # MA parameter
arma_process = ArmaProcess(ar, ma)
simulated_data = arma_process.generate_sample(nsample=100)

# Create a time series plot
plt.figure(figsize=(10, 4))
plt.plot(simulated_data)
plt.title('Simulated Time Series Data')
plt.xlabel('Time')
plt.ylabel('Value')
plt.grid(True)
plt.show()

# Plot the autocorrelation of the data
from statsmodels.graphics.tsaplots import plot_acf

plt.figure(figsize=(10, 4))
plot_acf(simulated_data, lags=20)
plt.title('Autocorrelation of the Data')
plt.xlabel('Lags')
plt.ylabel('Autocorrelation')
plt.grid(True)
plt.show()
```
