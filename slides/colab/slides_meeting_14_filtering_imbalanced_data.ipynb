```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate simulated data
np.random.seed(0)  # For reproducibility
n_vars = 25
data = np.random.randn(100, n_vars)  # 100 rows, 3 variables

# Step 2: Create a DataFrame
cols = [f'V{_}' for _ in range(n_vars)]
df = pd.DataFrame(data, columns=cols)

# Step 3: Calculate the Correlation Matrix
corr_matrix = df.corr()

# Step 4: Visualize the Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=n_vars<5, cmap='coolwarm')
plt.title('Correlation Matrix')
plt.show()
```

```python
from sklearn.datasets import make_blobs
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate isotropic Gaussian blobs
data, _ = make_blobs(n_samples=100, centers=3, n_features=4, random_state=42)

# Step 2: Introduce correlation by applying a linear transformation
transformation_matrix = np.array([[0.6, -0.3, -0.6, .2],
                                   [0.4, 0.08, -0.4, .7],
                                   [0.03, 0.3, 0.9, .04],
                                  [0.1, -0.2, 0.9, .14]])
data_transformed = np.dot(data, transformation_matrix)

# Step 3: Create a DataFrame
df = pd.DataFrame(data_transformed, columns=['Feature_1', 'Feature_2', 'Feature_3', 'Feature_4'])

# Step 4: Calculate the Correlation Matrix
corr_matrix = df.corr()

# Step 5: Visualize the Correlation Matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('Correlation Matrix')
plt.show()
```
