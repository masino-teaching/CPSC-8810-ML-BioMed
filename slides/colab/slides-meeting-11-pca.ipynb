```python
import numpy as np
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

# Step 1: Simulate the data
# Generating a dataset with 300 samples and 5 features
np.random.seed(0)  # For reproducibility
X = np.random.rand(300, 5)

# Step 2: Apply PCA
# It's a good practice to standardize the data before applying PCA
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=2)  # We only keep the first two components
X_pca = pca.fit_transform(X_scaled)

# Step 3: Visualize the first two principal components
plt.figure(figsize=(8, 6))
plt.scatter(X_pca[:, 0], X_pca[:, 1], alpha=0.7)
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.title('PCA of Simulated Dataset')
plt.grid(True)
plt.show()
```

Create simulated regression data

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

# Step 1: Create simulated classification data
X, y = make_classification(n_samples=300, n_features=5, n_informative=5, n_redundant=0, n_clusters_per_class=1, class_sep=2, random_state=42)

# Step 2: Apply PCA to the predictive features
# Standardizing the features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Applying PCA to reduce dimensions to 2
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X_scaled)

# Step 3: Transform the predictive features using the first two principal components
# (This step is effectively done in the previous step by applying PCA transformation)

# Step 4: Plot the transformed features as a scatter plot
plt.figure(figsize=(8, 6))
colors = ['navy', 'turquoise']
target_names = ['Class 0', 'Class 1']

for color, i, target_name in zip(colors, [0, 1], target_names):
    plt.scatter(X_pca[y == i, 0], X_pca[y == i, 1], color=color, alpha=.8, lw=2, label=target_name)

plt.legend(loc='best', shadow=False, scatterpoints=1)
plt.title('PCA of Simulated Dataset')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.grid(True)
plt.show()
```

```python
import numpy as np
from sklearn.datasets import make_classification
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D  # Import the 3D plotting tool

# Step 1: Create simulated classification data
X, y = make_classification(n_samples=300, n_features=5, n_informative=5, n_redundant=0, n_clusters_per_class=1, class_sep=2, random_state=42)

# Step 2 and 3: Apply PCA to the predictive features and transform using the first three principal components
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

pca = PCA(n_components=3)  # Keeping the first three components
X_pca = pca.fit_transform(X_scaled)

# Step 4: Plot the transformed features as a 3D scatter plot
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')  # Create a 3D plotting object

colors = ['navy', 'turquoise']
target_names = ['Class 0', 'Class 1']

for color, i, target_name in zip(colors, [0, 1], target_names):
    ax.scatter(X_pca[y == i, 0], X_pca[y == i, 1], X_pca[y == i, 2], color=color, alpha=.8, lw=2, label=target_name)

ax.legend(loc='best', shadow=False, scatterpoints=1)
ax.set_title('3D PCA of Simulated Dataset')
ax.set_xlabel('Principal Component 1')
ax.set_ylabel('Principal Component 2')
ax.set_zlabel('Principal Component 3')
plt.show()
```

```python
from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

# Assuming X_scaled is your standardized dataset from before
pca_full = PCA()  # Not specifying n_components uses all components
pca_full.fit(X_scaled)

explained_variance = pca_full.explained_variance_ratio_

plt.figure(figsize=(8, 6))
plt.bar(range(1, len(explained_variance) + 1), explained_variance, alpha=0.6, align='center', label='Individual explained variance')
plt.plot(range(1, len(explained_variance) + 1), np.cumsum(explained_variance),marker='x', label='Cumulative explained variance')
plt.ylabel('Explained variance ratio')
plt.xlabel('Principal component index')
plt.legend(loc='best')
plt.tight_layout()
plt.title('Scree Plot of the Principal Components')
plt.show()

```
